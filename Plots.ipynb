```python
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
X_LABEL = "Number of Pixels in Image"
X_LABEL2 = "Squared Root of Number of Pixels in Image"
```

```python
df_h = pd.read_csv('data_huffman.csv')
df_h = df_h.round(4)
df_h = df_h.sort_values('image_size')
```

```python
df_h
```

```python
img_size = df_h.iloc[:, 1].values
cp_h = df_h.iloc[:, 2].values
nobpp_h = df_h.iloc[:, 3].values
ct_h = df_h.iloc[:, 4].values
cm_h = df_h.iloc[:, 5].values
dt_h = df_h.iloc[:, 6].values
dm_h = df_h.iloc[:, 7].values
```

```python
df_ih = pd.read_csv('data_improved_huffman.csv')
df_ih = df_ih.round(4)
df_ih = df_ih.sort_values('image_size')
```

```python
df_ih
```

```python
img_size = df_ih.iloc[:, 1].values
cp_ih = df_ih.iloc[:, 2].values
nobpp_ih = df_ih.iloc[:, 3].values
ct_ih = df_ih.iloc[:, 4].values
cm_ih = df_ih.iloc[:, 5].values
dt_ih = df_ih.iloc[:, 6].values
dm_ih = df_ih.iloc[:, 7].values
```

```python
df_a = pd.read_csv('data_arithmetic.csv')
df_a = df_a.round(4)
df_a = df_a.sort_values('image_size')
```

```python
df_a
```

```python
img_size = df_a.iloc[:, 1].values
cp_a = df_a.iloc[:, 2].values
nobpp_a = df_a.iloc[:, 3].values
ct_a = df_a.iloc[:, 4].values
cm_a = df_a.iloc[:, 5].values
dt_a = df_a.iloc[:, 6].values
dm_a = df_a.iloc[:, 7].values
```

```python
print(df_a.columns.tolist())
```

```python
sqrt_img_size = np.sqrt(img_size)
```

```python
# Compression Time

plt.close()
plt.plot(img_size,ct_h,color='red', lw = 2,  label='Huffman')
plt.plot(img_size,ct_a,color='green', lw = 2, label='Arithmetic')
plt.plot(img_size,ct_ih,color='blue', lw = 2, label='Improved Huffman')
plt.xlabel(X_LABEL)
plt.ylabel("Compression Time (in seconds)")
plt.legend()
plt.savefig('./figures/ct.jpg', format = 'jpg')
plt.show()
```

```python
# Decompression Time

plt.close()
plt.plot(img_size,dt_h,color='red', lw = 2,  label='Huffman')
plt.plot(img_size,dt_a,color='green', lw = 2, label='Arithmetic')
plt.plot(img_size,dt_ih,color='blue', lw = 2, label='Improved Huffman')
plt.xlabel(X_LABEL)
plt.ylabel("Decompression Time (in seconds)")
plt.legend()
plt.savefig('./figures/dt.jpg', format = 'jpg')
plt.show()
```

```python
# compression memory

plt.close()
plt.plot(img_size,cm_h,color='red', lw = 2,  label='Huffman')
plt.plot(img_size,cm_a,color='green', lw = 2, label='Arithmetic')
plt.plot(img_size,cm_ih,color='blue', lw = 2, label='Improved Huffman')
plt.xlabel(X_LABEL)
plt.ylabel("Compression Memory (in Kbs)")
plt.legend()
plt.savefig('./figures/cm.jpg', format = 'jpg')
plt.show()
```

```python
# decompression memory

plt.close()
plt.plot(img_size,dm_h,color='red', lw = 2,  label='Huffman')
plt.plot(img_size,dm_a,color='green', lw = 2, label='Arithmetic')
plt.plot(img_size,dm_ih,color='blue', lw = 2, label='Improved Huffman')
plt.xlabel(X_LABEL)
plt.ylabel("Deompression Memory (in Kbs)")
plt.legend()
plt.savefig('./figures/dm.jpg', format = 'jpg')
plt.show()
```

```python
# compression percentage

plt.close()
plt.plot(sqrt_img_size,cp_h,color='red', lw = 2,  label='Huffman')
plt.plot(sqrt_img_size,cp_a,color='green', lw = 2, label='Arithmetic')
plt.plot(sqrt_img_size,cp_ih,color='blue', lw = 2, label='Improved Huffman')
plt.xlabel(X_LABEL2)
plt.ylabel("Compression Percentage")
plt.legend()
plt.savefig('./figures/cp.jpg', format = 'jpg')
plt.show()
```

```python
# compression percentage

plt.close()
plt.plot(sqrt_img_size,nobpp_h,color='red', lw = 2,  label='Huffman')
plt.plot(sqrt_img_size,nobpp_a,color='green', lw = 2, label='Arithmetic')
plt.plot(sqrt_img_size,nobpp_ih,color='blue', lw = 2, label='Improved Huffman')
plt.xlabel(X_LABEL2)
plt.ylabel("Number of bits per pixel")
plt.savefig('./figures/nobpp.jpg', format = 'jpg')
plt.show()
```
